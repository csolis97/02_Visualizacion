## Gráficos basicos   
A lo largo del .ipynb se generarán visualizaciones utilizando graficos basicos 

```python
# Librerias de python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np 
```

### Line Chart o gráfico de línea

Uno de los gráficos más comunes, su uso sugerido es para mostrar tendencias o evoluciones de datos temporales.  
Tambien se podria usar para comparar multiples series de datos, en donde el N de series es crucial para entender

```python
# Variables de ejemplo para visualizar 
X = np.linspace(-5, 5, num=50) # Crear un array 
Y = np.cos(X)
Z = np.sin(X)
# Grafico
fig, ax = plt.subplots()
ax.set_title("Gráfico de línea", x=0.15, fontsize=16)
ax.plot(X, Y,"3--",label="cos", color="orange")# agrega los datos al axis
ax.plot(X, Z,"4--",label="seno", color="blue")# agrega los datos al axis
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.minorticks_on()
ax.legend()
fig.set_layout_engine()
```

Se pueden generar distintas tipos de combinacione "linea"-"marcador":
| Estilo de Línea | Descripción              | Marcador | Descripción            |
|-----------------|--------------------------|----------|------------------------|
| `'-'`           | Línea sólida             | `'.'`    | Punto                  |
| `'--'`          | Línea discontinua        | `','`    | Pixel                  |
| `'-.'`          | Línea discontinua y punto| `'o'`    | Círculo                |
| `':'`           | Línea punteada           | `'v'`    | Triángulo hacia abajo  |
|                 |                          | `'^'`    | Triángulo hacia arriba |
|                 |                          | `'<'`    | Triángulo hacia la izquierda |
|                 |                          | `'>'`    | Triángulo hacia la derecha  |
|                 |                          | `'1'`    | Triángulo hacia abajo (otro estilo) |
|                 |                          | `'2'`    | Triángulo hacia arriba (otro estilo) |
|                 |                          | `'3'`    | Triángulo hacia la izquierda (otro estilo) |
|                 |                          | `'4'`    | Triángulo hacia la derecha (otro estilo) |
|                 |                          | `'s'`    | Cuadrado               |
|                 |                          | `'p'`    | Pentágono              |
|                 |                          | `'*'`    | Estrella               |
|                 |                          | `'h'`    | Hexágono 1             |
|                 |                          | `'H'`    | Hexágono 2             |
|                 |                          | `'+'`    | Signo más              |
|                 |                          | `'x'`    | X                      |
|                 |                          | `'D'`    | Diamante               |
|                 |                          | `'d'`    | Diamante pequeño       |
|                 |                          | `'|'`    | Línea vertical         |
|                 |                          | `'_'`    | Línea horizontal       |




```python
fig, ax = plt.subplots(1, 3, figsize=(12, 4))
# Primer axis
ax[0].plot(X, Y,"3--",label="cos", color="orange")
ax[0].set_title("Coseno de x",  fontsize=12)
ax[0].set_xlabel("x")
ax[0].set_ylabel("y")
ax[0].legend()
# Segundo axis
ax[1].plot(X, Z,"4--",label="seno", color="blue")
ax[1].set_title("Seno de x")
ax[1].set_xlabel("x")
ax[1].set_ylabel("y")
ax[1].legend()
# Tercer axis
ax[2].plot(X, Y,"3--",label="cos", color="orange")
ax[2].set_title("Coseno y seno de x",  fontsize=12)
ax[2].set_xlabel("x")
ax[2].set_ylabel("y")
ax[2].plot(X, Z,"4--",label="seno", color="blue")
ax[2].set_title("Seno de x")
ax[2].set_xlabel("x")
ax[2].set_ylabel("y")
ax[2].legend()
fig.tight_layout()
```
