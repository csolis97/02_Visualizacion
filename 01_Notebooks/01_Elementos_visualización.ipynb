#### **02: Elementos de una visualización** 
Se detalla cada componente de una visualización en `matplotlib`, a la cual, se puede acceder a cada elementos mediante distintos comandos

```python
from IPython.display import Image
Image("https://matplotlib.org/stable/_images/sphx_glr_anatomy_001.png")
```

| Componente       | Descripción                                                                 |
|------------------|-----------------------------------------------------------------------------|
| Figure (Figura)  | El contenedor global de la gráfica.                                         |
| Axis (Ejes)      | Los ejes son las líneas que delimitan el área de la gráfica y sobre las cuales se representan los datos. |
| Title (Título)   | El título de la gráfica.                                                    |
| Labels (Etiquetas) | Las etiquetas de los ejes X e Y.                                           |
| Legend (Leyenda) | La leyenda que describe los datos representados.                            |
| Grid (Cuadrícula) | Las líneas de la cuadrícula que ayudan a visualizar mejor los datos.        |
| Subplot (Subgráfica) | Una gráfica dentro de una figura que permite múltiples visualizaciones.  |
| Tick (Marca)     | Las marcas en los ejes que indican los valores de los datos.                |
| Spine (Espina)   | Las líneas que representan los bordes de los ejes en la gráfica.            |


#### **02: Comandos para acceder a los distintos elementos de una visualización** 
Los siguientes bloques de codigo contendran los comandos para acceder a los distintos elementos de uia visualización

```python
# Variables de ejemplo para visualizar
X = np.linspace(-5, 5, num=50) # Crear un array 
Y = np.cos(X)
```

```python
# Figura inicial
fig, ax = plt.subplots() # crea una figura (fig) y un conjunto de  axis (ax)
fig2, ax1 = plt.subplots(2, 1)
Otro_ejemplo = plt.subplots(3, 2, figsize=(12, 4)) # no es necesario que el nombre de la variable sea fig, ax ; pero facilita la ejecución de comandos orientados al axis
```

para este caso se crearon tres figueras, las cuales contenian disntintas cantidades de axis (ejes) donde se almacenarian nuestras visualizaciones, utilizando ademas, el atributo figsize para controlar el tamaño de la visualización



PSI; `plt.subplot` soporta distintos atributos, cómo;

| Atributo       | Descripción                                                                 |
|----------------|-----------------------------------------------------------------------------|
| sharex         | Controla si los subplots comparten el eje X.                                |
| sharey         | Controla si los subplots comparten el eje Y.                                |
| squeeze        | Si es True, elimina las dimensiones adicionales de la matriz de ejes.       |
| figsize        | Tamaño de la figura en pulgadas (ancho, alto).                              |
| dpi            | Resolución de la figura en puntos por pulgada.                              |
| facecolor      | Color de fondo de la figura.                                                |
| edgecolor      | Color del borde de la figura.                                               |
| num            | Número o etiqueta de la figura.                                             |
| tight_layout   | Ajusta automáticamente los subplots para que se ajusten a la figura.        |

```python
# Uso de atributos de plt.subplot
# Crear una figura con un tamaño personalizado y una resolución específica
fig, ax = plt.subplots(1, 2,figsize=(4, 2), dpi=100, facecolor='lightgray', edgecolor='blue', tight_layout=True)
fig.patch.set_linewidth(3) # Ajustamos el ancho de "edge" para efectos de visualizar el comando
```

Continuemos personalizando los elementos de una figura

```python
variable = np.average([1, 2, 3, 4, 5])
# Figura y axis
fig, (ax, ax2) = plt.subplots(1,2)
# Titulo
fig.suptitle("Titulo de fig", color='blue', fontsize=16, x=0.05)
ax.set_title("Titulo del axis", alpha=0.5, fontsize=8,x=0.2)
ax2.set_title("Titulo del axis 2", alpha=0.5,fontsize=8, x=0.8)
ax2.set_facecolor(color='lightgray')
# Ejes
ax.set_xlabel("Eje X")
ax.set_ylabel("Eje Y")
ax.set_xlim(0, 6)
ax.set_ylim(0, variable)
ax2.set_xlim(0, 6)
ax2.set_ylim(0, variable)
# Ticks
ax.set_yticks([0, 0.1, 0.3, 0.6, variable]) # Uso especificos de ticks
ax.set_yticklabels([0, 'a', 0.3, 'c' , variable]) 
ax.set_xticks([i for i in range(0, 7)])
ax.tick_params(axis='x', labelcolor='blue')
ax.tick_params(axis='x', which='major', length=0)
ax2.minorticks_on()
ax2.tick_params(axis='x', which='both', labelcolor='grey', color='grey')
ax2.tick_params(axis='y', which='both', labelcolor='grey', color='grey')
# Legenda
ax.legend(loc='upper right')
# Grilla
ax.grid(alpha= 0.2, color='blue')
# Spine
for spine in ax2.spines.values():
    spine.set_edgecolor('lightgray')
ax.spines['bottom'].set_edgecolor('blue') # se podria ajustar solo uno: ‘top’, ‘bottom’, ‘left’, y ‘right’

fig.tight_layout() # Introducimos comando para ajustar automáticamente los subplots dentro de la fig optimizando el espacio
```
